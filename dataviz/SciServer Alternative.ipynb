## 1.1 Signing Up

The main SciServer website has a "Login to SciServer" button. Click that. From here, click the "Create a new account" link and follow the steps.

You will receive an email that asks you to verify your email address. You must do so before signing in for the first time.

**Note:** Make sure you write down your password and don't forget it! If you type it wrong, you will be locked out of your account for a few minutes.*

## 1.2 Signing In

After signing in, you'll be presented with icons for various apps the site supports. We'll be using *Compute* (and *not* Compute Jobs).

## 1.3 Containers

The first time you sign in, you'll need to create a new Container. A container is basically a virtual computer that runs all of your software and keeps all of the packages you've installed up to date. Your files are independent of your image, so don't worry about where you keep things yet.

You'll need to do the following:

* Name the container whatever you like.
* Set the Compute Image to "SciServer Essentials."
* Check the "All" box next to "User volumes."

Then click create. Once it's made, you can click on a container's name to open it.

## 1.4 Exploring JupyterLab

Once you've opened your container, click the link that says "Switch to JupyterLab", which will provide you with a nice interface showing both your file browser on the left, and any notebooks or other files you're looking at on the right. Currently, the right panel just shows the launcher, which lets you make new files of different kinds.

When you log in, there should be two folders: Storage and Temporary. If not, log out and log back in about 15 minutes later. SciServer isn't done setting up your user space yet.

**Make sure all of your files are in /Storage/username/persistent/ or one of its subfolders.** Anything not in one of these will be deleted when the Container shuts down.

Here are some basic things to do:

1. Upload a file. You can do this by dragging and dropping a file from your computer to the file browser on the left. You can try getting and uploading the [*cereal.csv*](https://www.kaggle.com/crawford/80-cereals)$^2$ dataset we'll be using in Chapter 3. You can also upload a file using the up arrow icon (which should be right below the "Settings" menu).

2. Use the launcher to make a textfile and take some notes.

3. Bring up a new launcher (using the + in the upper left) and then make an R 3.6.2 notebook. Try typing the following in the cell that is presented to you, and then type Shift+Enter. If everything works, you will see the graph shown below.

> x <- 1:100/15<br>
> plot(x, sin(x) )

4. Right-click on your notebook and rename it something else. Then close its tab, right-click and delete it.

That's all for now.
